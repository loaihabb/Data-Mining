```python
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_profit_in_sector = data.nlargest(1, "Profits")
    df = df.append(highest_profit_in_sector) # bunun yerine concat kullanmamız lazım append kaldırılıcak diye 
df
```

```python
new_df = pd.DataFrame(columns=fortune.columns)
new_df
```

```python
for sector, data in sectors:
    highest_profit_in_sector = data.nlargest(1, "Profits") #her sectorde en çok kazanca sahip şirket 
    new_df = pd.concat([new_df, highest_profit_in_sector])
new_df
```

```python
cities = fortune.groupby("Location")
```

```python
new_df = pd.DataFrame(columns=fortune.columns)
new_df
```
