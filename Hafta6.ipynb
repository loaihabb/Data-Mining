```python
emp_df = pd.read_csv("Desktop/resources/employees.csv", parse_dates=["Start Date", "Last Login Time"])# date time'e çevirmek için ilk yol
#emp_df["Start Date"] = pd.to_datetime(emp_df["Start Date"]) # bu da ikinci yol , ikisiyle aynı sonuca çıkabiliriz (parse_date)
#emp_df["Last Login Time"] = pd.to_datetime(emp_df["Last Login Time"])
emp_df["Senior Management"] = emp_df["Senior Management"].astype(bool)
emp_df["Gender"] = emp_df["Gender"].astype("category")
emp_df # böyle doğrudan herşey çevirmiş oluruz (yukarıdakinin hepsi)
```

```python
emp_df[emp_df["Gender"] == "Male"] # Genderi male olanları bana getir demek 
emp_df[emp_df["Team"] == "Finance"]
```

```python
cond = emp_df["Gender"] == "Male"
emp_df[cond] # üsttekiyle aynı,  bu daha yardımcı olabilir 
```

```python
emp_df[emp_df["Senior Management"]] # direkt true olanları getirir
```

```python
emp_df[emp_df["Team"] != "Finance"] # eşit değildir,  eşit olmayanları getirir
```

```python
emp_df[emp_df["Salary"] > 50000] # 50 binden büyük olanları getirir
```

```python
emp_df[emp_df["Start Date"] > "1985-01-01" ] # bu tarihten sonra gelenleri getirir 
```

```python
emp_df[(emp_df["Team"] == "Marketing") & (emp_df["Start Date"] > "1990-01-01")] # ve için & kullanırız
 
#ikisi aynı sonucu getiriyor 
#aşağıdaki daha iyi

cond1 = emp_df["Team"] == "Marketing"
cond2 = emp_df["Start Date"] > "1990-01-01"
emp_df[cond1 & cond2]
```

```python
cond1 = emp_df["Team"] == "Marketing"
cond2 = emp_df["Start Date"] > "1990-01-01"
emp_df[cond1 & cond2]
```

```python
emp_df[(emp_df["Team"] == "Marketing") | (emp_df["Start Date"] > "1990-01-01")]  # veya için | dik çizgi kullanırız
```

```python
cond1 = emp_df["First Name"] == "Robert"
cond2 = emp_df["Start Date"] > "2016-06-01"
cond3 = emp_df["Team"] == "Client Services"

emp_df[cond1 & cond3 | cond2 ] # ve, veya böyle kullanılır 
# bi tane dosya 
```

```python
emp_df[emp_df["Team"].isin(["Marketing", "Finance" , "Client Services"])] # veya ile yazmadan böyle yazdık daha iyi
```

```python
emp_df[emp_df["Team"].isnull()] # null olanları getirir 
```

```python
emp_df[emp_df["Team"].notnull()] # null olmayanları getirir 
```

```python
emp_df[emp_df["Salary"].between(40000, 50000)] # iki değerin arasında olur # right = sağ tarafı dahil et sol tarafı etme
# normalde both olur 
emp_df[emp_df["Salary"].between(40000, 50000, inclusive="both")] # ikisini dahil et, neither= hiç birini dahil etme
```

```python
emp_df[emp_df["Start Date"].between("1985-01-01", "1990-01-01")] # bu iki tarih arasında işe başlayanlar
```

```python
emp_df[emp_df["Last Login Time"].between("08:30 AM", "12:00 PM")] #bu iki saatin arasında giriş yapanlar 
```

```python
emp_df.sort_values("First Name", inplace=True) # First Name'e göre sırala, değişiklikleri kaydet demek 
emp_df
```

```python
emp_df["First Name"].duplicated() # tekrar değerlendirmek 
emp_df[emp_df["First Name"].duplicated(keep="last")] # tekrarlanan isim, sonuncuyu saklar 
emp_df[emp_df["First Name"].duplicated(keep=False)] # saklanmıyor  
emp_df[~emp_df["First Name"].duplicated(keep=False)] # tekrarlanmayan isimleri getirir ~ deği

```

```python
emp_df.drop_duplicates() # tekrarlanan kayıtları siler, tam aynı olan
emp_df.drop_duplicates(subset=["First Name"]) # ilkini saklayıp tekrar olanları sildi
emp_df.drop_duplicates(subset=["First Name"], keep=False) # keep ile hangilarinin saklanacağını veriririz
emp_df.drop_duplicates(subset=["First Name", "Team"]) #aynı isiminde ve aynı takımında olanların ilkini tutup diğerlerini siler
```

```python
emp_df["Gender"].unique() # kaç tane unique değer varmış diye gösterir
emp_df["Gender"].nunique(dropna=False) # dropna yazmazsak nanları görmez 2 olur sonuç
len(emp_df["Gender"].unique()) # yukarıdakiyle aynı ama dropna=False iken
```
