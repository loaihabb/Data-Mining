```python
df_city = pd.read_csv("Desktop/resources/chicago.csv")
df_city.head()

```

```python
df_city.info() # bellekte ne kadar yer tuttuğunu görebiliriz
```

```python
df_city["Position Title"].str.title()# Kelimelerin sadece baş harflerini büyük gerisini küçük yapar
df_city["Position Title"].str.lower()#Bütün harfleri küçük yapar
df_city["Position Title"].str.upper() #Bütün harfleri büyük yapar
df_city["Position Title"].str.len() #sayılarını verir
```

```python
df_city.dropna(how="all",inplace=True) # Satırın hepsi NaN ise Satırı komple siliyor(how="all", değişiklikleri kaydetmek için)
```

```python
df_city["Department"].str.replace("MGMNT","MANAGEMENT") # department alanındaki MGMNT yazılan yerlere MANAGMENT yazar
```

```python
df_city["Employee Annual Salary"] = df_city["Employee Annual Salary"].str.replace("$","",regex=False).astype(float)
df_city
#Değişiklikleri kaydettik 
#Dolar Regular Expression'den olduğu için regex kullanmamız lazım 
#float'a çevirdik 
```

```python
df_city[df_city["Department"].str.startswith("WATER")] # WATER ile başlayan kelimeler 
df_city[df_city["Department"].str.endswith("WATER")] # WATER ile biten kelimeler 
df_city[df_city["Department"].str.contains("WATER")] # içinde WATER kelimesi olan kelimeler 
```

```python
df_city["Name"].str.lstrip() # String'in sol tarafındaki boşluğu siler
df_city["Name"].str.rstrip() # String'in sağ tarafındaki boşluğu siler
df_city["Name"].str.strip() ## String'in hem sağ hem de sol tarafındaki boşluğu siler(kelimelerin arasındaki boşlukları silmiyor)
```

```python
df_city["Name"].str.split("+") # kelimeleri Virgül ile ayırıyor

# str.get(0) -> ilk kelimeyi ayırıyor
# str.title -> kelimelerin baş harfini büyük yapar
#value.counts -> bir kelimenin kaç defa kaullandığını gösterir
df_city["Position Title"].str.split("-")
```
