```python
df_bigmac = pd.read_csv("Desktop/resources/bigmac.csv", parse_dates=["Date"],
                       index_col=["Date", "Country"]) # böyle de multiIndex tanımlayabiliriz istersek ilk baştan
df_bigmac.head()
```

```python
df_bigmac.index.get_level_values(0)#hangi leveldeki? outer 0,
df_bigmac.index.get_level_values("Date") # bu da aynı 

df_bigmac.index.get_level_values(1)#hangi leveldeki?  inner 1...
df_bigmac.index.get_level_values("Country")
```

```python
df_bigmac.index.set_names(["Date", "Location"])#indexlerin ismini değiştirebiliriz, kalıcı da olabilir,birinci bu
df_bigmac.index.set_names(names="Day", level=0) # böyle bir tanesini değiştirebiliriz istersek 
df_bigmac.index.set_names(names="Day", level="Date") #Date indexini Day yap
```

```python
df_bigmac.sort_index(ascending=False) # default olarak asending yapar, Büyükten küçüğe
df_bigmac.sort_index(ascending=[True, False])

df_bigmac.sort_index(level=0) # küçükten büyüğe ilk indexte sıralama yap
df_bigmac.sort_index(level="Date")

df_bigmac.sort_index(level=1)
df_bigmac.sort_index(level="Country")

```

```python
df_bigmac.loc[('2016-01-01', 'Argentina')] # ,"Price in US Dollars"
```

```python
df_bigmac.iloc[0]
df_bigmac.iloc[[0,2 ,4, 5, 250]]
```

```python
new_bigmac = df_bigmac.transpose() #satrırları sütun sütunları satır yapar
new_bigmac
```

```python
new_bigmac.loc[("Price in US Dollars",), ("2010-01-01", "Australia")] # seri, virgül var fiye
new_bigmac.loc[("Price in US Dollars"), ("2010-01-01", "Australia")] #değer virgül yoksa değer
new_bigmac.loc[("Price in US Dollars"), ("2010-01-01")]


new_bigmac.sort_index(inplace=True)
new_bigmac.loc[("Price in US Dollars",),
               ("2010-01-01", "Australia"):("2010-01-01", "Costa Rica")]

new_bigmac.iloc[0, 1:5]
```

```python
df_bigmac.swaplevel()
df_bigmac.swaplevel(0,1)
df_bigmac.swaplevel("Date", "Country") # inplace yok üstüne yazamamız gerekiyor
#Indexlerin yerlerini değiştiriyor  


```

```python
df_sales = pd.read_csv("Desktop/resources/salesmen.csv", parse_dates=["Date"])
df_sales.head()
```

```python
df_sales["Salesman"].unique() # kaç farklı değer var onu gösterir
```

```python
df_sales["Salesman"].value_counts() # bu değerden kaç tane geçtiğini gösterir
```

```python
df_foods = pd.read_csv("Desktop/resources/foods.csv")
df_foods
```

```python
df_foods.pivot_table(index="Gender", values="Spend", aggfunc="sum") # Gender'e göre harcamaları topla ve göster
# max , mean , min , std 

df_foods.pivot_table(index=["Gender", "Item"], values="Spend", aggfunc="count") # birden fazla index olabilir

df_foods.pivot_table(index=["Gender", "Item"], values="Spend", aggfunc="count",
                     columns=["Frequency", "City"])
```

```python
pd.pivot_table(df_foods, index="Gender", values="Spend", aggfunc="sum") # pandasın kendi pivot table'i, 
#Female 25963.33 , Male 24106.04
```
